(2) Limpieza de datos - Formatear timestamp de los viajes y eliminar los incongruentes

```python
df_bikeshare_test_clean = df_bikeshare_test_dropna.copy()

# Convertir start_time y end_time a datetime, detectando formatos mixtos
df_bikeshare_test_clean["start_time"] = pd.to_datetime(df_bikeshare_test_clean["start_time"], format='mixed')
df_bikeshare_test_clean["end_time"] = pd.to_datetime(df_bikeshare_test_clean["end_time"], format='mixed')

# Filtrar datos para conservar donde sea: end_time > start_time
df_bikeshare_test_clean = df_bikeshare_test_clean[df_bikeshare_test_clean["end_time"] > df_bikeshare_test_clean["start_time"]]
```

(3) Feature engineering - Agregar característica de "distance_km"

```python
import numpy as np


# Definir la función de distancia de Haversine
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Radio de la Tierra en km
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2.0) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2.0) ** 2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    return R * c



# Calcular la distancia y agregarla al dataframe
df_bikeshare_test_clean["distance_km"] = haversine(
    df_bikeshare_test_clean["start_lat"],
    df_bikeshare_test_clean["start_lon"],
    df_bikeshare_test_clean["end_lat"],
    df_bikeshare_test_clean["end_lon"]
)

df_bikeshare_test_clean.head()
```

(3) Feature engineering - Agregar columnas nuevas para extraer características del timestamp

```python
# Definir estaciones del año en Los Ángeles, California
def get_season(month):
    if month in [3, 4, 5]:
        return 1  # Primavera
    elif month in [6, 7, 8]:
        return 2  # Verano
    elif month in [9, 10, 11]:
        return 3  # Otoño
    else:
        return 4  # Invierno


# Extraer información relevante
df_bikeshare_test_clean["start_hour"] = df_bikeshare_test_clean["start_time"].dt.hour
df_bikeshare_test_clean["start_weekday"] = df_bikeshare_test_clean["start_time"].dt.weekday
df_bikeshare_test_clean["start_month"] = df_bikeshare_test_clean["start_time"].dt.month
df_bikeshare_test_clean["start_year"] = df_bikeshare_test_clean["start_time"].dt.year

# Agregar la estación del año
df_bikeshare_test_clean['start_season'] = df_bikeshare_test_clean['start_month'].apply(get_season)

# Definir si es día laboral o fin de semana
df_bikeshare_test_clean['start_day_type'] = df_bikeshare_test_clean['start_weekday'].apply(lambda x: 1 if x in [5, 6] else 0)  # 0 = Laboral, 1 = Fin de semana

df_bikeshare_test_clean.head()
```
